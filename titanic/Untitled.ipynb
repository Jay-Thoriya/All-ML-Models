```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df = df.drop(df[['Name','Ticket','Cabin']],axis=1)
```

```python
df.head()
```

```python
df['Sex'] = df['Sex'].astype('category').cat.codes
```

```python
df.head()
```

```python
df["Embarked"].unique()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.replace({"Embarked":{'S':0, 'C':1, 'Q':2,'nan':0}})
```

```python
df.head()
```

```python
df["Embarked"].unique()
```

```python
df.set_index('PassengerId', inplace=True)
```

```python
df.head()
```

```python
df['Age'].isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
df.head()
```

```python
for i in df['Age']:
    if i < 18:
        df['Age'] = df['Age'].replace(i,0)
    elif i >= 18 and i < 40:
        df['Age'] = df['Age'].replace(i,1)
    elif i >= 40 and i < 60:
        df['Age'] = df['Age'].replace(i,2)
    else:
        df['Age'] = df['Age'].replace(i,3)
```

```python
df.describe()
```

```python
for i in df['Fare']:
    if i < 128:
        df['Fare'] = df['Fare'].replace(i,0)
    elif i >= 128 and i < 256:
        df['Fare'] = df['Fare'].replace(i,1)
    elif i >= 256 and i < 384:
        df['Fare'] = df['Fare'].replace(i,2)
    else:
        df['Fare'] = df['Fare'].replace(i,3)
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df.drop('Survived', axis=1)
```

```python
X.head()
```

```python
y = df['Survived']
```

```python
y.head()
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report

dtree = DecisionTreeClassifier()

dtree.fit(X_train, Y_train)
```

```python
prediction = dtree.predict(X_test)
```

```python
print("classification_report : ",classification_report(Y_test,prediction))
print("confusion_matrix : ",confusion_matrix(Y_test,prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(X_train, Y_train)
```

```python
predictions_rfc = rfc.predict(X_test)
```

```python
print("classification_report : ",classification_report(Y_test,predictions_rfc))
print("confusion_matrix : ",confusion_matrix(Y_test,predictions_rfc))
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()

xgb.fit(X_train, Y_train)
```

```python
predictions_xgb = xgb.predict(X_test)
```

```python
print("classification_report : ",classification_report(Y_test,predictions_xgb))
print("confusion_matrix : ",confusion_matrix(Y_test,predictions_xgb))
```

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(X_train, Y_train)

predictions_svc = svc.predict(X_test)

print("classification_report : ",classification_report(Y_test,predictions_svc))
print("confusion_matrix : ",confusion_matrix(Y_test,predictions_svc))
```
