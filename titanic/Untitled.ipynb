```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df = df.drop(df[['Name','Ticket','Cabin']],axis=1)
```

```python
df.head()
```

```python
df['Sex'] = df['Sex'].astype('category').cat.codes
```

```python
df.head()
```

```python
df["Embarked"].unique()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.replace({"Embarked":{'S':0, 'C':1, 'Q':2,'nan':0}})
```

```python
df.head()
```

```python
df["Embarked"].unique()
```

```python
df.set_index('PassengerId', inplace=True)
```

```python
df.head()
```

```python
df['Age'].isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
df.head()
```

```python
for i in df['Age']:
    if i < 18:
        df['Age'] = df['Age'].replace(i,0)
    elif i >= 18 and i < 40:
        df['Age'] = df['Age'].replace(i,1)
    elif i >= 40 and i < 60:
        df['Age'] = df['Age'].replace(i,2)
    else:
        df['Age'] = df['Age'].replace(i,3)
```
